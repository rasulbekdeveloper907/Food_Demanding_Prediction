# Raqamli va skewness > 1 bo‘lgan ustunlarga fokus qilamiz:

```python
import numpy as np

# log1p = log(x + 1), bu 0 ga teng bo‘lgan qiymatlar uchun xavfsizroq
df['num_orders_log'] = np.log1p(df['num_orders'])
df['price_diff_log'] = np.log1p(df['price_diff'])

# Optional: checkout_price, base_price
df['checkout_price_log'] = np.log1p(df['checkout_price'])
df['base_price_log'] = np.log1p(df['base_price'])
```

# price_diff_ratio ustunidagi NaN muammo

```python
# Avval base_price = 0 bo'lganlarni tekshiramiz
print(df[df['base_price'] == 0])

# Ehtimol, bu yerda bo'sh yoki xatolik bor — buni ehtiyotkorlik bilan hal qilish kerak
# Masalan, price_diff_ratio ustunini qayta hisoblaymiz:
df['price_diff_ratio'] = df['price_diff'] / (df['base_price'] + 1e-5)
```

# Skewnessdan keyin yana normalizatsiya (scaling)

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_cols = ['num_orders_log', 'price_diff_log', 'checkout_price_log', 'base_price_log']
df[scaled_cols] = scaler.fit_transform(df[scaled_cols])
```

```python
skewness = df.skew(numeric_only=True)
print(skewness)
```

```python
df['num_orders_sqrt'] = np.sqrt(df['num_orders'])
```

# price_diff_ratio ustuni: Skewness -698 bo‘lishi — bu yerda ehtimol base_price = 0 bo‘lib, katta manfiy qiymatlar kelib chiqqan.

```python
# Avval bazaviy 0 bo'lganlar sonini ko‘raylik
print(df[df['base_price'] == 0].shape[0])

# So'ng NaN yoki 0 bo'lishdan qochish uchun denominatorga kichik son qo‘shamiz:
df['price_diff_ratio'] = df['price_diff'] / (df['base_price'] + 1e-5)

# Yoki: limit qo‘yamiz
df['price_diff_ratio'] = df['price_diff_ratio'].clip(-10, 10)
```
