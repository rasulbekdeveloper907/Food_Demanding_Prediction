```python
# Istalgan tekst kiritilsin
# 1) kichik harflarda qilib olsin  
# 2) oradagi spacelarni olib tashlasin
# 3) teskari qilib yozib bersin 
```

```python
# 1) kichik harflarda qilib olsin  
def to_lower(text):
    return text.lower()
```

```python
# 2) oradagi spacelarni olib tashlasin
def remove_space(text):
    return text.replace(" ", "")
```

```python
# 3) teskari qilib yozib bersin 
def teskari(text):
    return text[::-1]
```

```python
pipeline = [to_lower, remove_space, teskari]
data = "Hello Ai"
for step in pipeline:
    data=step(data)
print(data)
```

```python
pipeline = [remove_space, teskari]
data = "Hello Ai"
for step in pipeline:
    data=step(data)
print(data)
```

```python
pipeline = [remove_space]
data = "Hello Ai"
for step in pipeline:
    data=step(data)
print(data)
```

```python
pipeline = [to_lower, remove_space]
data = "Hello Ai"
for step in pipeline:
    data=step(data)
print(data)
```

# Kichik harflarga o‘tkazish → Belgilar sonini chiqarish

```python
def to_lower(text):
    return text.lower()

def count_chars(text):
    return len(text)

pipeline = [to_lower, count_chars]

data = "mening ishim pipeline"

for step in pipeline:
    data = step(data)

print("Belgilar soni:", data)  
```

# So‘zni teskari qilish → Har bir harfni alohida chiqarish

```python
def reverse_text(text):
    return text[::-1]

def print_each_char(text):
    for ch in text:
        print(ch)

pipeline = [reverse_text, print_each_char]

data = "mushukvoy"

for step in pipeline:
    data = step(data)
```

# Faqat unli harflarni ajratib olish

```python
def extract_vowels(text):
    vowels = "aeiouAEIOU"
    return ''.join([ch for ch in text if ch in vowels])

pipeline = [extract_vowels]

data = "olamshumul yangiliklar bor "

for step in pipeline:
    data = step(data)

print("Unli harflar:", data)  
```

# So‘zlarni alifbo tartibida saralash

```python
def sort_words(text):
    words = text.split()
    words.sort()
    return ' '.join(words)

pipeline = [sort_words]

data = "salom bu yerda nima gaplar"

for step in pipeline:
    data = step(data)

print("Tartiblangan so‘zlar:", data)
```
